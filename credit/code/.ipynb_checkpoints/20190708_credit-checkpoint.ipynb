```python
# Read in data
credit = pd.read_excel('../input/default of credit card clients.xls', header =1)
```

```python
#Change columns order 
colnames = credit.columns.tolist()
colnames =  ['ID', 'LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE', 'PAY_6', 'PAY_5', 'PAY_4', 'PAY_3', 'PAY_2', 'PAY_0', 'BILL_AMT6', 'BILL_AMT5', 'BILL_AMT4', 'BILL_AMT3', 'BILL_AMT2', 'BILL_AMT1', 'PAY_AMT6', 'PAY_AMT5', 'PAY_AMT4', 'PAY_AMT3', 'PAY_AMT2', 'PAY_AMT1', 'default payment next month']
credit = credit[colnames]

#rename some columns so their name makes more sense
credit.columns = ['ID', 'CREDIT', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE', 'PAYSTAT_APR', 'PAYSTAT_MAY', 'PAYSTAT_JUNE', 'PAYSTAT_JULY', 'PAYSTAT_AUG', 'PAYSTAT_SEP', 'CUR_BIL_APR', 'CUR_BIL_MAY', 'CUR_BIL_JUNE', 'CUR_BIL_JULY', 'CUR_BIL_AUG', 'CUR_BIL_SEP', 'PAID_APR', 'PAID_MAY', 'PAID_JUNE', 'PAID_JULY', 'PAID_AUG', 'PAID_SEP', 'DEFAULT']
```

```python
plt.hist(credit['CREDIT'])
plt.show()
```

```python
credit.columns = credit.columns.str.strip()
print (credit.columns.tolist())
```

```python
#Change values to meaningful
credit['DEFAULT'] = credit['DEFAULT'].replace({0: 'Performing', 
                                               1: 'Default'})
credit['SEX'] = credit['SEX'].replace({1: 'Male', 
                                       2: 'Female'})
credit['MARRIAGE'] = credit['MARRIAGE'].replace({0: 'Other', 
                                                    1: 'Single', 
                                                    2: 'Married', 
                                                    3: 'Other'})
credit['EDUCATION'] = credit['EDUCATION'].replace({0: 'Other', 
                                                   1: 'Graduate School', 
                                                   2: 'University',
                                                   3: 'High School',
                                                   4: 'Other',
                                                   5: 'Other',
                                                   6: 'Other'})
```

```python
#Change type of categorical variables from int to cat
for col in ['SEX', 
            'EDUCATION', 
            'MARRIAGE', 
            'AGE', 
            'PAYSTAT_SEP', 'PAYSTAT_AUG', 'PAYSTAT_JULY', 'PAYSTAT_JUNE', 'PAYSTAT_MAY', 'PAYSTAT_APR', 
            'DEFAULT']:
    credit[col] = credit[col].astype('category')
```

```python
#Make a variable that states whether the person has defaulted in the past
credit['DEFAULTED_BEFORE'] = (np.where((credit['PAYSTAT_APR'] == 1) |
                                       (credit['PAYSTAT_MAY'] == 1) |
                                       (credit['PAYSTAT_JUNE'] == 1) |
                                       (credit['PAYSTAT_JULY'] == 1) |
                                       (credit['PAYSTAT_AUG'] == 1) |
                                       (credit['PAYSTAT_SEP'] == 1)
                                       , 1 , 0))
```

```python
#Correlation matrix
corrMat = credit.corr()
print(corrMat)
```

```python
#convert all categories into numerical variables so they can be proberly used to model with
credit.SEX = pd.CategoricalIndex(credit.SEX)
credit.EDUCATION = pd.CategoricalIndex(credit.EDUCATION)
credit.MARRIAGE = pd.CategoricalIndex(credit.MARRIAGE)
credit.DEFAULT = pd.CategoricalIndex(credit.DEFAULT)
credit.DEFAULTED_BEFORE = pd.CategoricalIndex(credit.DEFAULTED_BEFORE)

credit['SEX'] = credit.SEX.cat.codes
credit['EDUCATION'] = credit.EDUCATION.cat.codes
credit['MARRIAGE'] = credit.MARRIAGE.cat.codes
credit['DEFAULT'] = credit.DEFAULT.cat.codes
credit['DEFAULTED_BEFORE'] = credit.DEFAULTED_BEFORE.cat.codes
```

```python
credit.head()
```

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(credit, credit['DEFAULT']):
    #print(train_index, test_index)
    strat_train_set = credit.loc[train_index]
    strat_test_set = credit.loc[test_index]
```

```python
#Verify that a stratified split has been applied
credit['DEFAULT'].value_counts() / len(credit)
strat_test_set['DEFAULT'].value_counts() / len(strat_test_set)
strat_train_set['DEFAULT'].value_counts() / len(strat_train_set)
```

```python
credit_features = strat_train_set.drop('DEFAULT', axis=1)
credit_labels = strat_train_set['DEFAULT'].copy()
test_features = strat_test_set.drop('DEFAULT', axis=1)
test_labels = strat_test_set['DEFAULT'].copy()
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(credit_features, credit_labels)
```

```python
# Make variable for the imported models
modelSVC = SVC()
modelRF = RandomForestClassifier()
modelLR = LogisticRegression()
modelKNN = KNeighborsClassifier
```

```python
# Do a Kfold cross validation on the training data for k = 5

knn = KNeighborsClassifier(n_neighbors = 5)
CVscores = cross_val_score(knn, credit_features, credit_labels, cv = 10, scoring = "accuracy")
print(CVscores.mean())
```

```python
# Do a Kfold cross validation on the training data to select optimal K 
k_range = range(2,10)
k_scores= []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    CVscores = cross_val_score(knn, credit_features, credit_labels, cv = 10, scoring = "accuracy")
    k_scores.append(CVscores.mean())
print(k_scores)
```

```python
plt.plot(k_range, k_scores)
plt.xlabel('Number of Neighbors')
plt.ylabel('Cross validation score')
```

```python
# Check classification accuracy using the top K that was retrieved from the Knn cross-validation
knn = KNeighborsClassifier(n_neighbors = k_range[((k_scores.index(max(k_scores))))])
knn.fit(credit_features, credit_labels)
y_pred = knn.predict(test_features)
print("Accuracy =", metrics.accuracy_score(test_labels, y_pred))
print("Cohen's Kappa =", cohen_kappa_score(test_labels, y_pred))
print("Recall =", recall_score(test_labels, y_pred))
print("F1 score =", f1_score(test_labels, y_pred))
print("ROC area under curve =", roc_auc_score(test_labels, y_pred))
```

```python
# Do a Kfold cross validation on the training data for a logistic regression
logregression = LogisticRegression(solver='liblinear')
CVscores = cross_val_score(logregression, credit_features, credit_labels, cv = 10, scoring = "accuracy")
print(CVscores.mean())
```

```python
# Check classification accuracy using the logregression
logregression.fit(credit_features, credit_labels)
y_pred = logregression.predict(test_features)
print("Accuracy =", metrics.accuracy_score(test_labels, y_pred))
print("Cohen's Kappa =", cohen_kappa_score(test_labels, y_pred))
print("Recall =", recall_score(test_labels, y_pred))
print("F1 score =", f1_score(test_labels, y_pred))
print("ROC area under curve =", roc_auc_score(test_labels, y_pred))
```

```python
#train using a Support vector machine, a random forest and a logistic regression, 
modelSVC.fit(credit_features, credit_labels)
modelRF.fit(credit_features, credit_labels)
```

```python
#Print cross-validation scores 
print(cross_val_score(modelSVC, credit_features, credit_labels)) 
print(cross_val_score(modelRF, credit_features, credit_labels)) 

#Print model validation scores 
print(modelSVC.score(credit_features, credit_labels))
print(modelRF.score(credit_features, credit_labels))
```

```python
# Check classification accuracy of using the Support Vector Machine
y_pred = modelSVC.predict(test_features)
print("Accuracy =", metrics.accuracy_score(test_labels, y_pred))
print("Cohen's Kappa =", cohen_kappa_score(test_labels, y_pred))
print("Recall =", recall_score(test_labels, y_pred))
print("F1 score =", f1_score(test_labels, y_pred))
print("ROC area under curve =", roc_auc_score(test_labels, y_pred))
print(confusion_matrix(test_labels, y_pred))
```

```python
# Check classification of accuracy using the Random forest
y_pred = modelRF.predict(test_features)
print("Accuracy =", metrics.accuracy_score(test_labels, y_pred))
print("Cohen's Kappa =", cohen_kappa_score(test_labels, y_pred))
print("Recall =", recall_score(test_labels, y_pred))
print("F1 score =", f1_score(test_labels, y_pred))
print("ROC area under curve =", roc_auc_score(test_labels, y_pred))
print(confusion_matrix(test_labels, y_pred))
```

```python
importances = modelRF.feature_importances_
std = np.std([tree.feature_importances_ for tree in modelRF.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(credit_features.shape[1]):
    print("%d. feature %s (%f)" % (f + 1, credit_features.columns[f], importances[indices[f]]))

# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(credit_features.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(credit_features.shape[1]), indices)
plt.xlim([-1, credit_features.shape[1]])
plt.show()
```
